```python
import numpy as np
import matplotlib.pyplot as plt

from task_c import Solution_c
```

```python
#import the result
solution = Solution_c()
results = solution.stepCalculation()
```

```python
#print the diagram
x = np.linspace(1,10000, num=10000, endpoint=True, dtype=int)
plt.figure(figsize=(40,10)) 
plt.scatter(x, results, alpha=0.6)
#add max and min values
max_indx = np.argmax(results)
min_indx = np.argmin(results)
plt.plot(max_indx, results[max_indx], 'rs')
plt.plot(min_indx, results[min_indx], 'rs')
show_max='  ['+str(max_indx+1)+' '+str(results[max_indx])+']'
show_min='  ['+str(min_indx+1)+' '+str(results[min_indx])+']'
plt.annotate(show_max, xy=(max_indx,results[max_indx]))
plt.annotate(show_min, xy=(min_indx,results[min_indx]))

plt.title('The steps for [1, 10000]')
plt.xlabel('start number')
plt.ylabel('steps')
plt.show()
```
